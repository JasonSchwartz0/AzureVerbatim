```python
# Initialize new columns
df['Count_of_Matched_Phrases'] = 0
df['Removal_Flag'] = ''
df['Checker_Feedback'] = ''
```

```python
# Iterate over each column
for column in columns_to_check:
    # Iterate over each paragraph in the column
    for index, paragraph in df[column].items():  # Use items() instead of iteritems()
        count = 0
        if pd.notnull(paragraph):
            for phrase in premade_list:
                if pd.notnull(phrase):
                    count += str(paragraph).lower().count(str(phrase).lower())
        df.at[index, 'Count_of_Matched_Phrases'] += count
```

```python
# Define a function to map values
def determine_removal_flag(count):
    if count in (3, 4, 5):
        return "Flag"
    elif count >= 6:
        return "Fraudulent"
    else:
        return "Keep"
```

```python
# Apply the function to create the 'Removal_Flag' column
df['Removal_Flag'] = df['Count_of_Matched_Phrases'].apply(determine_removal_flag)
```

```python
# Define the check_for_colon function
def check_for_colon(value):
    return ':' in str(value)

# Check for missing responses in columns_to_check
missing_responses = df[columns_to_check].isnull().any(axis=1)

# Update 'Removal_Flag' and 'Checker_Feedback' columns based on missing responses
df.loc[missing_responses, ['Removal_Flag', 'Checker_Feedback']] = ['Fraudulent', 'Missing Responses']

# Mark duplicates except for the first occurrence based on 'IPAddress' column
df['is_duplicate_ip'] = df.duplicated(subset='IPAddress', keep='first')

# Mark duplicates except for the first occurrence based on responses
duplicate_responses = df.duplicated(subset=columns_to_check, keep=False)
df['is_duplicate_response'] = duplicate_responses

# Mark duplicates except for the first occurrence based on colon presence
df['has_colon'] = df[columns_to_check].apply(lambda x: x.map(check_for_colon)).any(axis=1)

# Update 'Removal_Flag' and 'Checker_Feedback' columns based on duplicate IP address
df.loc[df['is_duplicate_ip'], ['Removal_Flag', 'Checker_Feedback']] = ['Fraudulent', 'Duplicate IP Address']

# Update 'Removal_Flag' and 'Checker_Feedback' columns based on duplicate responses
df.loc[df['is_duplicate_response'], ['Removal_Flag', 'Checker_Feedback']] = ['Fraudulent', 'Duplicate Response']

# Update 'Removal_Flag' and 'Checker_Feedback' columns based on colon presence
df.loc[df['has_colon'], ['Removal_Flag', 'Checker_Feedback']] = ['Fraudulent', 'Used a colon']
df.loc[df['Checker_Feedback'] == 'Used a colon', 'Likely_AI'] = 'Yes'

# Update 'Likely_AI' column for Count_of_Matched_Phrases >= 6
df.loc[df['Count_of_Matched_Phrases'] >= 6, 'Likely_AI'] = 'Yes'

# Drop the temporary columns
df.drop(columns=['is_duplicate_ip', 'is_duplicate_response', 'has_colon'], inplace=True)

# Function to check if a response has only one word
def is_single_word(response):
    return isinstance(response, str) and len(response.split()) == 1

# Check for single-word responses in specified columns
single_word_responses = df[columns_to_check].apply(lambda x: x.map(is_single_word)).any(axis=1)

# Replace NaN values in 'Likely_AI' column with 'No'
df['Likely_AI'].fillna('No', inplace=True)
df['Likely_AI'] = df['Likely_AI'].replace({np.nan: 'No'})

# Update 'Removal_Flag' and 'Checker_Feedback' columns for single-word responses
df.loc[single_word_responses, ['Removal_Flag', 'Checker_Feedback']] = ['Fraudulent', 'Vague Response']
```
